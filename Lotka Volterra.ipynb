```python
import numpy as np
import matplotlib.pyplot as plt
```

## Lotka-Volterra

```python
r=3
x=np.linspace(0,10,25)
y=np.linspace(0,6,25)
x,y=np.meshgrid(x,y)
u=x*(1-y)
v=y*(x-r)

fixed=([0,r],[0,1])
seed=([0,6],[2,0],[4,2],[4,4],[3,3],[2.5,1],[2.5,1.5],[3,5],[3,6],[9,4.5],[8.5,4],[2.5,3.5],[10,5])

plt.figure(figsize=(20,15))
plt.plot(fixed[0],fixed[1],'bo',markersize=10)
plt.style.use('default')
color = np.sqrt(u**2+v**2)
plt.quiver(x,y,u,v,color,alpha=0.5)
plt.streamplot(x,y,u,v, density=50,color='red',arrowsize=3,start_points=seed)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
cbar=plt.colorbar()
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(18)
plt.show()
```

```python
R=3
dt=0.001
n=25000
t=np.linspace(0, (n+1)*dt, n+2)
X=np.zeros(n+2)
Y=np.zeros(n+2)
X[0]=6
Y[0]=3

for i in range(n+1):
    X[i+1] = X[i] + X[i]*(1-Y[i])*dt
    Y[i+1] = Y[i] + Y[i]*(X[i]-r)*dt

plt.figure(figsize=(12,8))  
plt.style.use('seaborn-whitegrid')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Time',fontsize=15)
plt.ylabel('Characteristic Population Density',fontsize=15)
plt.plot(t,X,'r',label="Prey",linewidth=2)
plt.plot(t,Y,'b',label='Predator',linewidth=2)
plt.legend(loc='upper right',frameon=True,fontsize=15)
plt.show()
```

## Rosensweig

### Stable Spiral, $\tau^2-4\Delta<0$

```python
r=2
k=5
x=np.linspace(0,6,25)
y=np.linspace(0,2,25)
x,y=np.meshgrid(x,y)
u=x-(x*y)-(x*x/k)
v=(-r*y)+(x*y)

fixed=([0,k,r],[0,0,1-(r/k)])
seed=([0,2],[2,0],[2,1],[5.1,0],[4,1],[3,1],[5,1.5],[5,1],[6,1.5])

plt.figure(figsize=(20,15))
plt.plot(fixed[0],fixed[1],'bo',markersize=10)

color = np.sqrt(u**2+v**2)
q=plt.quiver(x,y,u,v,color,alpha=0.5)
X=np.linspace(2.5,5,25)
Y=(1-r+k)+((-1+r-k)*X/k)
plt.style.use('default')
plt.plot(X,Y,linestyle='dashed',color='blue',alpha=0.5,linewidth=3)
plt.streamplot(x,y,u,v, density=100,color='red',arrowsize=3,start_points=seed)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
cbar=plt.colorbar()
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(18)
plt.show()
```

```python
R=2
K=5
dt=0.001
n=25000
t=np.linspace(0, (n+1)*dt, n+2)
X=np.zeros(n+2)
Y=np.zeros(n+2)
X[0]=6
Y[0]=3

for i in range(n+1):
    X[i+1] = X[i] + X[i]*(1-Y[i]-X[i]/K)*dt
    Y[i+1] = Y[i] + Y[i]*(X[i]-R)*dt
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,8))  

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.xlabel('Time',fontsize=15)
plt.ylabel('Characteristic Population Density',fontsize=15)
plt.plot(t,X,'r',label="Prey",linewidth=2)
plt.plot(t,Y,'b',label='Predator',linewidth=2)
plt.legend(loc='upper right',frameon=True,fontsize=15)
plt.show()
```

### Stable Node, $\tau^2-4\Delta>0$

```python
r=3
k=3.1
x=np.linspace(0,6,25)
y=np.linspace(0,0.25,25)
x,y=np.meshgrid(x,y)
u=x-(x*y)-(x*x/k)
v=(-r*y)+(x*y)

fixed=([0,k,r],[0,0,1-(r/k)])
seed=([0,0.25],[2,0],[4,0],[4,0.05],[4,0.1],[4,0.15],[4,0.2],[1.0,0.2],[1.5,0.2],[2.0,0.2],[2.5,0.2],[4,0.02])

plt.figure(figsize=(20,15))
plt.plot(fixed[0],fixed[1],'bo',markersize=10)
plt.style.use('default')
color = np.sqrt(u**2+v**2)
q=plt.quiver(x,y,u,v,color,alpha=0.5)

X=np.linspace(k-0.7,k,25)
Y=-(r-k-1)+((r-k-1)*X/k)
plt.plot(X,Y,linestyle='dashed',color='blue',alpha=0.5,linewidth=3)

eigen=0.5*(-(r/k)+np.sqrt(((r**2)/(k**2))-(4*r)+(4*(r**2)/k)))
W=np.linspace(r-0.78,r+0.125)
C=((eigen+(r/k))*r)+(k*(1-(r/k)))
Z=C/k-((r/k)+eigen)*W/k
plt.plot(W,Z,linestyle='dashed',color='green',alpha=0.5,linewidth=3)

plt.streamplot(x,y,u,v, density=200,color='red',arrowsize=3,start_points=seed)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
cbar=plt.colorbar()
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(18)
plt.show()
```

### Extinction, $\rho>K$

```python
r=5
k=2
x=np.linspace(0,6,25)
y=np.linspace(0,0.25,25)
x,y=np.meshgrid(x,y)
u=x-(x*y)-(x*x/k)
v=(-r*y)+(x*y)

fixed=([0,k],[0,0])
seed=([0,0.25],[1,0],[4,0],[4,0.05],[4,0.1],[4,0.15],[4,0.2],[0.5,0.2],[1.0,0.2],[1.5,0.2],[2.0,0.2],[2.5,0.2],[4,0.02])

plt.figure(figsize=(20,15))
plt.plot(fixed[0],fixed[1],'bo',markersize=10)

color = np.sqrt(u**2+v**2)
q=plt.quiver(x,y,u,v,color,alpha=0.5)

X=np.linspace(k,k+0.25,25)
Y=-(r-k-1)+((r-k-1)*X/k)
plt.plot(X,Y,linestyle='dashed',color='blue',alpha=0.5,linewidth=3)

plt.style.use('default')
plt.streamplot(x,y,u,v, density=100,color='red',arrowsize=3,start_points=seed)
plt.xticks(fontsize=18)
plt.yticks(fontsize=18)
cbar=plt.colorbar()
for t in cbar.ax.get_yticklabels():
     t.set_fontsize(18)
plt.show()
```
